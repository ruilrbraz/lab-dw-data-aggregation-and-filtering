```python
import pandas as pd

url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv"
df = pd.read_csv(url)

df.info()
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
df.columns = [col.lower().replace(' ', '_') for col in df.columns]
df
```

```python
df['effective_to_date'] = pd.to_datetime(df['effective_to_date'], format='%m/%d/%y')
```

```python
median_value = df['months_since_last_claim'].median()

df['months_since_last_claim'] = df['months_since_last_claim'].fillna(median_value)
```

```python
mode_value = df['state'].mode()[0]

df['state'] = df['state'].fillna(mode_value)
```

```python
df = df.drop(columns=['vehicle_type'])
```

```python
df['number_of_open_complaints'] = df['number_of_open_complaints'].fillna(df['number_of_open_complaints'].median())
df['response'] = df['response'].fillna(df['response'].mode()[0])
df['vehicle_class'] = df['vehicle_class'].fillna(df['vehicle_class'].mode()[0])
df['vehicle_size'] = df['vehicle_size'].fillna(df['vehicle_size'].mode()[0])
```

```python
df.isna().sum()
```

```python
#1. Create a new DataFrame that only includes customers who:
# have a low total_claim_amount (e.g., below $1,000),
# have a response "Yes" to the last marketing campaign.

# Create the two conditions
low_claim_condition = df['total_claim_amount'] < 1000
yes_response_condition = df['response'] == 'Yes'

# Combine the conditions with '&' and apply them to the DataFrame
# Note the parentheses around each condition, which is important!
low_risk_customers_df = df[low_claim_condition & yes_response_condition]

print("New DataFrame of low-risk, engaged customers:")
low_risk_customers_df.head()
```

```python
#2. Using the original Dataframe, analyze:
# the average monthly_premium and/or customer lifetime value by policy_type and gender for customers who responded "Yes", and
# compare these insights to total_claim_amount patterns, and discuss which segments appear most profitable or low-risk for the company.

# First, filter for customers who responded "Yes"
yes_customers = df[df['response'] == 'Yes']

# Now, group by policy type and gender, and aggregate the key metrics
profitability_analysis = yes_customers.groupby(['policy_type', 'gender']).agg(
    average_monthly_premium=('monthly_premium_auto', 'mean'),
    average_clv=('customer_lifetime_value', 'mean'),
    average_claim_amount=('total_claim_amount', 'mean')
)

print("Profitability Analysis:")
profitability_analysis.round(2) # Round for readability
```

```python
#3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

# Get the number of customers in each state
state_counts = df['state'].value_counts()

# Filter the results to only include states with more than 500 customers
states_over_500 = state_counts[state_counts > 500]

print("States with over 500 customers:")
print(states_over_500)
```

```python
#4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

clv_by_segment = df.groupby(['education', 'gender'])['customer_lifetime_value'].agg(['max', 'min', 'median'])

print("CLV Analysis by Education and Gender:")
clv_by_segment.round(2)
```

```python
# Convert the column to datetime objects
df['effective_to_date'] = pd.to_datetime(df['effective_to_date'])

# Extract the month number and create a new 'month' column
df['month'] = df['effective_to_date'].dt.month
```

```python
#5. The marketing team wants to analyze the number of policies sold by state and month. 
# Present the data in a table where the months are arranged as columns and the states are arranged as rows.

# Create the pivot table
# We add fill_value=0 to make sure any state/month combo with no sales shows a 0 instead of NaN
policies_pivot = df.pivot_table(
    index='state',
    columns='month',
    values='number_of_policies',
    aggfunc='sum',
    fill_value=0
)

print("Policies Sold by State and Month:")
policies_pivot.unstack()
```

```python
#6. Display a new DataFrame that contains the number of policies sold by month, by state, for the top 3 states with the highest number of policies sold.

# Create a 'total' column by summing the policies across the months (the columns)
policies_pivot['total_policies'] = policies_pivot.sum(axis=1)

# Sort the table by this new 'total' column and get the index (the state names) of the top 3
top_3_states = policies_pivot.sort_values(by='total_policies', ascending=False).head(3).index

# Use .loc to select only the rows for the top 3 states from our pivot table
top_3_df = policies_pivot.loc[top_3_states]

# Drop the temporary 'total_policies' column for a clean final table
top_3_df = top_3_df.drop(columns=['total_policies'])


print("\nTop 3 States by Number of Policies Sold:")
print(top_3_df)
```
